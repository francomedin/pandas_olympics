## pd: Agregué las medallas ganadas por mujeres con una tabla multiIndex para obtener más información

## Construir una tabla que muestre cuántas medallas obtuvieron los hombres en total en cada año que se realizó el evento.

```python
data_set_by_year_gender=data_set.groupby(by=['Gender','Year'])[['Medal']].count()
```

```python
data_set_by_year_gender['Change']=data_set_by_year_gender.diff()


```

```python
data_set_by_year_gender.iloc[9,1]=0
data_set_by_year_gender=data_set_by_year_gender.fillna(value=0)
```

```python
data_set_by_year_gender
```

```python
data_set_by_year_gender[['Medal Change %' , 'Change Difference %']]=data_set_by_year_gender.pct_change()
```

```python
data_set_by_year_gender=data_set_by_year_gender.fillna(value=0)
data_set_by_year_gender[['Medal Change %' , 'Change Difference %']]=np.round(data_set_by_year_gender[['Medal Change %' , 'Change Difference %']] *100,decimals=2)
```

```python
data_set_by_year_gender.iloc[9,1:4]=0
data_set_by_year_gender
```
